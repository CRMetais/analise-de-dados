```python
import pandas as pd
```

```python
# Ler a base de dados
df = pd.read_csv('../bases/BASE_COMMODITIES.csv')
```

```python
# Converter coluna 'data' para datetime
df['data'] = pd.to_datetime(df['data'])
```

```python
# Expandir a coluna de data em ano, mês e dia
df['ano'] = df['data'].dt.year
df['mes'] = df['data'].dt.month
df['dia'] = df['data'].dt.day
```

```python
# Filtrar apenas "Minerais, Minérios E Metais"
df_minerais = df[df['NO_COMMODITIES'] == 'Minerais, Minérios E Metais'].copy()
```

```python
# Manter apenas as colunas desejadas
colunas = ['NO_COMMODITIES', 'data', 'ano', 'mes', 'dia', 'var_mensal_calc']
df_minerais = df_minerais[colunas]
```

```python
# Limitar var_mensal_calc a 2 casas decimais e renomear coluna
df_minerais['var_mensal_calc'] = df_minerais['var_mensal_calc'].round(2)
df_minerais = df_minerais.rename(columns={'var_mensal_calc': 'Var_mensal'})
```

```python
# Salvar a nova base
df_minerais.to_csv('../bases/BASE_COMMODITIES_TRATADA.csv', index=False)

print(f"Base de Minerais criada com {len(df_minerais)} registros")
print(f"Colunas: {df_minerais.columns.tolist()}")
print(f"\nPrimeiras linhas:")
print(df_minerais.head())
```
